```python
import seaborn as sns

df = sns.load_dataset("iris")

print("Shape: ", df.shape)
print("Columns: ", df.columns)
print("Head: ",df.head(5))


```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data=df, 
x="sepal_length", 
y="sepal_width", 
hue="species"
)
plt.title("Comparison of Sepal Length and Width by Species")
plt.show


```

```python
sns.scatterplot(data = df,
                x = "petal_length",
                y = "sepal_length",
                hue = "species"
               )   
plt.title("Comparison of Petal Length and Sepal Length by Species")
plt.show()  
```

```python
sns.scatterplot(data = df,
                x = "petal_width",
                y = "sepal_width",
                hue = "species"
               )   
plt.title("Comparison of Petal Width and Sepal Width by Species")
plt.show()  
```

```python
plt.figure()
plt.hist(df["sepal_length"], bins=20)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency") 
plt.show()
```

```python
plt.figure()
plt.hist(df["species"])
plt.title("Distribution of Species")
plt.xlabel("Species")
plt.ylabel("Frequency") 
plt.show()
```

```python
plt.figure()
sns.boxplot(data = df,
            x = "species",
             y = "sepal_length"
             )
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species") 
plt.ylabel("Sepal Length (cm)") 
plt.show()
```

```python
plt.figure()
sns.boxplot(data = df,
            x = "species",
             y = "petal_length"
             )
plt.title("Box Plot of Petal Length by Species")
plt.xlabel("Species") 
plt.ylabel("Petal Length (cm)") 
plt.show()
```
